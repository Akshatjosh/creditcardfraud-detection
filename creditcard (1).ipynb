```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report,confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import xgboost as xgb
```

# New Section

```python
data = pd.read_csv('/content/creditcard.csv')
```

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
legit_sample = legit.sample(n=len(fraud), random_state=2)
data = pd.concat([legit_sample, fraud], axis=0)
```

```python
# Split data into features and target
X = data.drop(columns="Class", axis=1)
y = data["Class"]
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
logistic_model = LogisticRegression(max_iter=500)
```

```python
logistic_model.fit(X_train, y_train)
```

```python
logistic_train_pred = logistic_model.predict(X_train)
logistic_test_pred = logistic_model.predict(X_test)
```

```python
logistic_train_acc = accuracy_score(y_train, logistic_train_pred) * 100
logistic_test_acc = accuracy_score(y_test, logistic_test_pred) * 100
```

```python
logistic_conf_matrix = confusion_matrix(y_test, logistic_test_pred)
```

```python
print("Logistic Regression:")
print(f"Train Accuracy: {logistic_train_acc}")
print(f"Test Accuracy: {logistic_test_acc}")
print("Confusion Matrix:")
print(logistic_conf_matrix)
```

```python
decision_tree = DecisionTreeClassifier(random_state=2)
```

```python
decision_tree.fit(X_train, y_train)
```

```python
tree_train_pred = decision_tree.predict(X_train)
tree_test_pred = decision_tree.predict(X_test)
```

```python
tree_train_acc = accuracy_score(y_train, tree_train_pred) * 100
tree_test_acc = accuracy_score(y_test, tree_test_pred) * 100
```

```python
tree_conf_matrix = confusion_matrix(y_test, tree_test_pred)
```

```python
y_probs = decision_tree.predict_proba(X_test)[:, 1]
```

```python
dt_auc = roc_auc_score(y_test, y_probs)
print("Decision Tree AUC:", dt_auc)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Build DNN model
model = Sequential([
    Dense(32, input_dim=X_train.shape[1], activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train model
history = model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

# Evaluate model
y_pred = (model.predict(X_test_scaled) > 0.5).astype("int32")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
model = Sequential()
model.add(Dense(16, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['AUC'])

# 2. Model train karna
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# 3. Prediction probabilities lena
y_probs_keras = model.predict(X_test).ravel()
```

```python
keras_auc = roc_auc_score(y_test, y_probs_keras)
print("Keras Model AUC:", keras_auc)
```

```python
y_pred_keras = (y_probs_keras >= 0.5).astype(int)
cm_keras = confusion_matrix(y_test, y_pred_keras)
print("Confusion Matrix (Keras):\n", cm_keras)
```

```python
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

# 2. Train karo
xgb_model.fit(X_train, y_train)

# 3. Predict probabilities for AUC
y_probs_xgb = xgb_model.predict_proba(X_test)[:, 1]
xgb_auc = roc_auc_score(y_test, y_probs_xgb)
print("XGBoost AUC:", xgb_auc)
```

```python
# 4. Predict labels for Confusion Matrix
y_pred_xgb = xgb_model.predict(X_test)
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
print("Confusion Matrix (XGBoost):\n", cm_xgb)
```
